```python
def convert_range(x):
    try:
        text = str(x).split(" - ")
        if len(text) == 2:
            val = (float(text[0]) + float(text[1])) / 2
            return val
        else:
            return float(x)
    except:
        return None # Or an appropriate default value
        

# Apply the function to the dataframe column
df['total_sqft'] = df['total_sqft'].apply(convert_range)
```

```python
def remove_outlier_sqft(df):
    df_output=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        gen_df=subdf[(subdf['price_per_sqft']>(m-st)) & (subdf['price_per_sqft']<(m+st))]
        df_output=pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
df=remove_outlier_sqft(df)
df.describe()
```

```python
df.shape
```

```python
def bhk_outlier_remover(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_status={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_status[bhk]={
                'mean': np.mean(bhk_df['price_per_sqft']),
                'std': np.std(bhk_df['price_per_sqft']),
                'count': bhk_df.shape[0]
            }
        print(location,bhk_status)
        for bhk,bhk_df in location_df.groupby('bhk'):
            status=bhk_status.get(bhk-1)
            if status and status['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df['price_per_sqft']<(status['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
df=bhk_outlier_remover(df)
```

```python
X=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

<h3>Linear Regression</h3>

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),
    remainder='passthrough'
)
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred_lr)
print(f"R^2 Score: {r2}")
```

<h3>Applying Lasso Regression</h3>

```python
from sklearn.linear_model import Lasso
pipe=make_pipeline(column_trans,scaler,Lasso())
pipe.fit(X_train,y_train)
y_pred_lasso=pipe.predict(X_test)
```

```python
r2_lasso = r2_score(y_test, y_pred_lasso)
print(f"R^2 Score: {r2_lasso}")
```

<h3>Applying Ridge Regression</h3>

```python
from sklearn.linear_model import Ridge
pipe=make_pipeline(column_trans,scaler,Ridge())
pipe.fit(X_train,y_train)
y_pred_ridge=pipe.predict(X_test)
```

```python
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f"R^2 Score: {r2_ridge}")
```
